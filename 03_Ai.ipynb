```python
netflix = pd.read_csv('netflix_selena.csv')
# .loc[] : 데이터프레임의 행과 열의 이름을 사용
# 전체 행과 title과 release_year의 열을 출력
title_year = netflix.loc[:,['title','release_year']]
title_year.head()
```

```python
# .loc[] : 데이터프레임의 행과 열의 이름을 사용
# 0번째, 1번째 행과 title과 release_year의 열을 출력
title_year = netflix.loc[[0,1],['title','release_year']]
title_year.head()
```

```python
# .index : 행 이름 변경
netflix.index = ['A','B','C','D','E','F','G','H']
# .loc[] : 데이터프레임의 행과 열의 이름을 사용
# B, C 행과 title과 release_year의 열을 출력
title_year = netflix.loc[['B','C'],['title','release_year']]
title_year.head()
```

```python
netflix = pd.read_csv('netflix_selena.csv')
# .iloc[] : 데이터프레임의 행과 열 번호를 사용
# 3번째, 4번째 행과 1번째, 3번째 열을 출력
type_director = netflix.iloc[[3,4],[1,3]]
type_director.head()
```

```python
# .isin(values) : 각 요소가 values에 포함된 값인지 파악하여 True/False 반환
# type 변수의 값이 TV Show일 경우 True 반환, 아니면 False
netflix['type'].isin(['TV Show'])
```

```python
# 논리형 인덱싱 + isin() : 데이터의 특정 범위만 추출
# type 변수의 값이 TV Show일 경우만 추출하여 TV_Show 저장
TV_Show = netflix[netflix['type'].isin(['TV Show'])]
TV_Show.head()
```

```python
netflix = pd.read_csv('netflix_selena.csv')
# 결측치 확인 (1) - .isna().sum()
netflix.isna().sum()
```

```python
# 결측치 확인 (2) - .info()
netflix.info()
```

```python
# 결측치 확인 (3) - for문
for i in netflix.columns:
  missingValueRate = netflix[i].isna().sum() / len(netflix) * 100
  if missingValueRate > 0:
    print("{} null rate : {}%".format(i, round(missingValueRate, 2)))
```
