```python
import numpy as np
x = np.linspace(0,10,100)
y1 = x
y2 = 2 * x
y3 = 0.5 * x
y4 = 3 * x

plt.plot(x,y1, label='y=x')
plt.plot(x,y2, label='y=2x')
plt.plot(x,y3, label='y=0.5x')
plt.plot(x,y4, label='y=3x')

plt.legend(loc='upper left',ncol=2)

plt.title('Simple Linear Plot Example')
plt.xlabel('x')
plt.ylabel('y')

plt.show()
```

```python
plt.plot([1,2,3,4],[3,6,9,12])
plt.xlim([0,5])
plt.ylim([0,15])
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(1,10,100)
y = np.exp(x)

plt.subplot(2,2,1)
plt.plot(x,y)
plt.title('Linear Scale')

plt.subplot(2,2,2)
plt.plot(x,y)
plt.yscale('log')
plt.title('Logarithmic Scale (y-axis)')

plt.subplot(2,2,3)
plt.plot(x,y)
plt.xscale('log')
plt.title('Log-Linear Scale (x-axis)')

plt.subplot(2,2,4)
plt.plot(x,y)
plt.yscale('log')
plt.xscale('linear')
plt.title('Linear-logarithmic Scale')

plt.tight_layout()
plt.show()
```
