```python
# length, height, width
#  길이,   높이,  너비

import numpy as np

# 무게 타깃 (정답파일용)
perch_weight = np.array(
    [5.9, 32.0, 40.0, 51.5, 70.0, 100.0, 78.0, 80.0, 85.0, 85.0,
     110.0, 115.0, 125.0, 130.0, 120.0, 120.0, 130.0, 135.0, 110.0,
     130.0, 150.0, 145.0, 150.0, 170.0, 225.0, 145.0, 188.0, 180.0,
     197.0, 218.0, 300.0, 260.0, 265.0, 250.0, 250.0, 300.0, 320.0,
     514.0, 556.0, 840.0, 685.0, 700.0, 700.0, 690.0, 900.0, 650.0,
     820.0, 850.0, 900.0, 1015.0, 820.0, 1100.0, 1000.0, 1100.0,
     1000.0, 1000.0]
     )
```

```python
from sklearn.model_selection import train_test_split

train_input, test_input, train_target, test_target = train_test_split(
    perch_full, perch_weight, random_state=42)
# 훈련세트와 테스트세트로 나눔

print(train_input[0], train_target[0])
# 판다스에서 받은 데이터 [length, height, width] weight
#                          19.6    5.14    3.04   85.0
```

```python
print(train_input.shape, test_input.shape)
# 훈련용 42개의 농어와 3개의 특성
# 테스트 14개의 농어와 3개의 특성
```

```python
# 위 데이터를 사용해서 새로운 특성을 만든다
# 사이킷런 변환기 : 특성을 만들거나 전처리하는 클래스가 있다.
# 변환기(Transformer) : fit(), transform() 메서드가 있다

from sklearn.preprocessing import PolynomialFeatures # 특성변환기
# 전에는 제곱을 이용해서 특성을 변경하여 곡서은로 만들었음
# PolynomialFeatures 사이킷런에서 수학자들이 만든 공식을 사용

poly = PolynomialFeatures()
poly.fit([[2,3]])
# 2개의 특성을 가진 2와 3으로 이루어진 샘플 하나를 적용 테스트
print(poly.transform([[2,3]]))
# fit() 메서드와 transform() 메서드를 차례대로 실행(훈련을 해야 변환이 됨)
```

```python
# [[1. 2. 3. 4. 6. 9.]]
# 1 (bias)
# 2   # fit에 첫번째 값
# 3   # fit에 두번째 값
# 4   = 2 ** 2
# 6   = 2 * 3
# 9   = 3 ** 2

# PolynomialFeatures는 기본적으로 각 특성을 제곱한 항을 추가하고
# 특성끼리 서로 곱한 항을 추가함
# 2와 3을 각기 제곱한 4와 9가 추가되고, 2와 3을 곱한 6이 추가됨
# 무게 = a * 길이 + b * 높이 + c + 두께 + d + 1
# 선형 방정식의 절편을 항상 값이 1인 특성과 곱해지는 계수임 -> (길이, 높이, 두께, 1)
```

```python
poly = PolynomialFeatures(include_bias=False)
#                           1 삭제 (자동으로 절편 항은 무시하지만 알아두면 좋음)
poly.fit([[2,3]])
print(poly.transform([[2,3]])) # 특성에 1이 무조건 계산되지 않기 때문에 속도 향상
```

```python
poly.get_feature_names_out() # 특성이 어떻게 만들어 졌는지 확인
```

```python
# 실제로 훈련시작용 생성
poly = PolynomialFeatures(include_bias=False)
# 사이킷런의 선형 모델은 자동으로 bias를 추가하므로
# bias 특성을 만들 필요 없으므로 include_bias를 False로 지정

poly.fit(train_input)
train_poly = poly.transform(train_input)
print(train_poly.shape) # 42개의 요소에 9개의 특성 (3개의 특성을 공학으로 9개로 늘림)
# 무게, 높이, 두께를 다중 결합하여 특성을 만듬 (특성공학)
```

```python
poly.get_feature_names_out() # 특성이 어떻게 만들어 졌는지 확인
```

```python
# ['x0', 'x1', 'x2', 'x0^2', 'x0 x1', 'x0 x2', 'x1^2', 'x1 x2', x2^2']
# length,height,width,length^2
#                           length*height
#                                   length*width
#                                             height^2
#                                                     height*width
#                                                               width^2
```

```python
test_poly = poly.transform(test_input) # 테스트용 세트 변환
```

```python
# 다중 회귀 모델 훈련하기
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(train_poly, train_target)
print(lr.score(train_poly, train_target))
# 99.03% 의 아주 좋은 점수가 나옴 (물고기의 길이, 높이, 두께까지 사용한 결과)
```

```python
print(lr.score(test_poly, test_target))
# 97.14% 농어의 길이만을 사용했을 때에 있던 과소적합 문제는 나타나지 않음
```

```python
from sklearn.preprocessing import StandardScaler
# StandardScaler 싸이킷런 데이터 전처리 스케일 조정(스케일러) -> 표준점수로 변환
# 예전에는 공식으로 평균과 표준편차를 사용하라고 했는데 StandardScaler를 사용하면 끝

ss = StandardScaler()    # 훈련세트로 학습한 변환기를 사용해 테스트 세트까지 변환 필수!!
ss.fit(train_poly)

train_scaled = ss.transform(train_poly) # 트레이닝용
test_scaled = ss.transform(test_poly) # 테스트용
# 표준점수로 변환한 train_scaled, test_scaled 준비완료
# 평균과 표준편차로 공식을 적용할 필요 없음 (표준점수화 됨)
```

```python
# 선형회귀 모델에 규제를 추가한 모델은 2개 있음
# 릿지(ridge) : 계수를 제곱한 값을 기준으로 규제 (선호)
# 라쏘(lasso) : 계수의 절대값을 기준으로 규제를 적용 -> 0으로 제공 가능
```

```python
# 릿지회귀 : 모델 객체를 만들고 fit() -> score()
from sklearn.linear_model import Ridge

ridge = Ridge() # 릿지 클래스를 객체로 생성하고 변수 연결
ridge.fit(train_scaled, train_target) # fit로 훈련하고 score로 평가
print(ridge.score(train_scaled, train_target))
# 테스트 결과 좋은 성능임을 판단 할 수 있음 (모델 객체 생성 시 alpha 매개변수로 규제의 강도를 조절)
```

```python
# 릿지와 라쏘 모델을 사용할 때 규제의 양을 임의로 조절할 수 있음
# 모델 객체를 만들 때 alpha 매개값을 주면 된다.
# alpha 값을 크게 주면 강도가 세짐, 계수값을 더 줄이고 조금 더 과소적합되도록 유도하는 용도

# 적절한 alpha 값을 찾은 방법은 그래프로 그려 봐야한다.
import matplotlib.pyplot as plt

train_score = []    # 훈련 : 규제의 알파 농도에 다라서 결과 %를 기록 저장용
test_score = []     # 테스트 % 저장용
# alpha 값을 바꿀 때마다 score() 메서드의 결과 값을 저장할 리스트 생성
```

```python
alpha_list = [0.001, 0.01, 0.1, 1, 10, 100] # 0.001 ~ 100까지 10배씩 증가
for alpha in alpha_list:    # 리스트 반복용
  # 릿지 모델 만들기
  ridge = Ridge(alpha=alpha)
  # 릿지 모델 훈련
  ridge.fit(train_scaled, train_target)
  # 훈련 점수와 테스트 점수 저장
  train_score.append(ridge.score(train_scaled,train_target)) # 훈련점수 리스트 보관
  test_score.append(ridge.score(test_scaled,test_target)) # 테스트점수 리스트 보관
print(f"train_score {train_score}")
print(f"test_score {test_score}")
```

```python
# 위 결과를 차트로 보기!

# .log10(alpha_list) x축 값이 너무 촘촘하게 보일까봐 지수형식으로 표현
# 0,001 : -3, 0.01 : -2, 0.1 : -1
plt.plot(np.log10(alpha_list), train_score) # 파란색
plt.plot(np.log10(alpha_list), test_score) # 주황색
plt.xlabel('alpha')
plt.ylabel('R^2')
plt.show() # 0.1 일 때 제일 좋은 점수로 파악
```

```python
# 지금 기본값도 안정적이지만 알파값을 조절하면 더 좋은 값을 찾울 수 있겠다?
train_score = []
test_score = []

alpha_list = [0.001, 0.01, 0.1, 1, 10, 100]
for alpha in alpha_list:
  # 라쏘 모델 만들기
  lasso = Lasso(alpha=alpha, max_iter=100000) # 최대 반복횟수 (정확도를 위해)
  # 라쏘 모델은 최적의 게수를 찾기 위해 반복적인 계산을 수행

  # 라쏘 모델 훈련
  lasso.fit(train_scaled, train_target)

  # 훈련 점수와 테스트 점수를 위에 만든 리스트에 저장
  train_score.append(lasso.score(train_scaled, train_target))
  test_score.append(lasso.score(test_scaled, test_target))

# 라쏘 테스트 시 반복 계산을 하는데 지정한 반복횟수가 부족하면 경고 발생 ConvergenceWarning
```

```python
# 위에서 만든 리스트를 차트로 보자
plt.plot(np.log10(alpha_list),train_score) # 파란색
plt.plot(np.log10(alpha_list),test_score) # 주황색
plt.xlabel('alpha')
plt.ylabel('R^2')
plt.show()
# 왼쪽은 과대적합, 오른쪽은 정상 (제일 좋은 값은 1 => 10)
```
